1. check if one string is a rotation of another string: you have two strings, str1 and str2 and you need to return True if the first string is a rotation of the second string. Otherwise, return False.

```python
def is_rotation(str1, str2):                                                    
    if len(str1) != len(str2):                                                    
        return False                                                                
    str_rot=[]                                                                            
    for i in range(len(str1)):
        str_rot.append(str1[i:] + str1[:i])     
    if str2 in str_rot:
        return True
    else: 
        return False
is_rotation('yes','esy')
```

2. Check whether a string is Palindrome or not

```python
def is_palindrome(st):
    return st == st[::-1]
is_palindrome('soes')
```

```python
is_palindrome('hellolleh')
```

3. Check one string  one string is a permutation of another string

```python
def are_permutations(str1, str2): 
    return (len(str1) == len(str2) and sorted(str1) == sorted(str2))
are_permutations('hello','lloeh')
```

### How to Select Top N Rows with the Largest Values in a Column(s) in Pandas?

This can be done with the function **nlargest** also has an argument keep that allows one to deal with duplicate values. keep can take {‘first’, ‘last’, ‘all’}, where   

- first : prioritize the first occurrence(s)   
- last : prioritize the last occurrence(s)   
- all : does not drop any duplicates   

```python
import pandas as pd
data = {'department':['a','a','a','b','b','b','b','c'],'name':['andrew','zello','fan','brown','ethan','apple','dang','huynh'],'salary':[10,9,12,10,10, 9,6,9]}
df= pd.DataFrame(data)
df.head(8)
```

**DataFrame.nlargest**   
This method is equivalent to df.sort_values(columns, ascending=False).head(n) , but more performant.   

columns: labels
keep : {‘first’, ‘last’, ‘all’}, default ‘first’

    Where there are duplicate values:

        first : prioritize the first occurrence(s)
        last : prioritize the last occurrence(s)

        all : do not drop any duplicates, even it means
            selecting more than n items.



```python
df= df.groupby('department').apply(lambda x: x.sort_values(["salary"], ascending = False)).reset_index(drop=True)
df.groupby('department').head(1)
```

```python
data_url = "https://goo.gl/ioc2Td"
gapminder = pd.read_csv(data_url)
print(gapminder.head(3))
```

We will filter our dataframe to contain life Expectancy values per year using Pandas loc and string matching.

```python
lifeExp = gapminder.loc[:, gapminder.columns.str.contains('^life|^c')]
print(lifeExp.head(n=3))
```

```python
gapminder_tidy = lifeExp.melt(id_vars=["continent", "country"], 
                              var_name="year", 
                              value_name="lifeExp")
gapminder_tidy.head(n=3)
```

We get the gapminder data in long form, with each variable as separate columns. However, note that the year column contains the prefix “lifeExp_”. We can filter by

```python
gapminder_tidy.year = gapminder_tidy.year.str.replace("lifeExp_","")`
gapminder_tidy.head(3)
```

Use **wide_to_long** to convert wide to long format.  we specify “stubnames” to extract the prefix from column variable names.

```python
lifeExp_tidy = pd.wide_to_long(lifeExp, stubnames=['lifeExp'], i='country',
                 j='year', sep='_')
```

```python
lifeExp_tidy.head(3)
```

```python
lifeExp_tidy.unstack().reset_index()
```

We can use Pandas’ wide_to_long()’s argument stubnames to specify multiple variables that we want to reshape to long form. For example, to reshape all three variables over time in gapminder dataframe in wide form, we specify the prefixes with stubnames=[‘lifeExp’, ‘gdpPercap’,’pop’].

```python
gapminder.head(3)
```

```python
gapminder_long = pd.wide_to_long(gapminder,stubnames= ['lifeExp', 'gdpPercap','pop'], 
                 i='id',
                 j='year', sep='_')
gapminder_long.head()
```

Another example with melt function

```python
df = pd.DataFrame([[2, 4, 7, 8, 1, 3, 2013], [9, 2, 4, 5, 5, 6, 2014]], columns=['Amy', 'Bob', 'Carl', 'Chris', 'Ben', 'Other', 'Year'])
df
```

```python
d = {'A': ['Amy'], 'B': ['Bob', 'Ben'], 'C': ['Carl', 'Chris']}
```

```python
m = pd.melt(df, id_vars=['Year'], var_name='Name')
m
```

This has everything except Group. To get that, we need to reshape d a bit as well.

```python
d2 = {}
for k, v in d.items():
    for item in v:
        d2[item] = k
d2        
```

```python
m['Group'] = m['Name'].map(d2)
m
```

```python
mask = m['Name'] == 'Other'
m.loc[mask, 'Name'] = ''
m.loc[mask, 'Group'] = 'Other'
```
