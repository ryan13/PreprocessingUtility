**Transform** is a method of data frame combining with **groupby** to aggregate data. The similar of **tranform** and **agg** are both used to generate a summary of rows grouping; the difference is **transform** keep origin data (rows) before grouping while **agg** is not.  

```python
import pandas as pd

df = pd.read_excel("./Downloads/data/sales_transactions.xlsx")
df.head()
```

```python
df.shape
```

### In general, agg function will reduce rows by number of groupby feature. Let tally each order

```python
order_total = df.groupby('order')["ext price"].sum().rename("Order_Total").reset_index()
order_total
```

Let see the output of **agg** method

```python
df.groupby('order').agg({'ext price':'sum'}).rename(columns={'ext price':'Total'} ).reset_index()
```

We got only three rows

### To generate the full report with  % for each order, we merge back to original data.   
Note: Two data frame are difference in size. Let look again the original data

```python
df
```

```python
order_total = df.groupby('order')["ext price"].sum().rename("Order_Total").reset_index()
df_1 = df.merge(order_total)
df_1["Percent_of_Order"] = df_1["ext price"] / df_1["Order_Total"]
```

### we can see columns : Order_Total, and Percent_of_Order

```python
df_1
```

This certainly works but there are several steps needed to get the data combined in the manner we need.   

### Second Approach - Using Transform
Using the original data, let’s try using transform and groupby and see what we get:

```python
df.groupby('order')["ext price"].transform('sum')
```

We can see that the output has difference with previous agg function (both either agg method and aggregate method)

```python
df["Order_Total"] = df.groupby('order')["ext price"].transform('sum')
df["Percent_of_Order"] = df["ext price"] / df["Order_Total"
df
```
